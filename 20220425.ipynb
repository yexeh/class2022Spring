Autocorrelation

-자기 자신(auto)와의 상관관계를 알아보고자 할 때 사용

-분석값의 시차값(time lag)의 유사성

-반복되는 패턴, missing fundamental frequency를 찾는 수학적 도구로서 사용

-함수 분석용으로 signal processing에서 사용

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import librosa, librosa.display
import IPython.display as ipd
```

```python
import os
url = "https://raw.githubusercontent.com/hsnam95/class2022Spring/main/aeiou.wav"
os.system("curl " + url + " > aeiou.wav")

s, sr = librosa.load('aeiou.wav')
```

```python
s = librosa.util.normalize(s)
librosa.display.waveplot(s, sr)
ipd.Audio(s[7000:12000], rate=sr)
```

```python
sm.graphics.tsa.plot_acf(s, lags=50, use_vlines=True)
```
